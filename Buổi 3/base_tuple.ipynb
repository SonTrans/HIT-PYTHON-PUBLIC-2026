**Tuple**

```python
tuple1 = (1,2, 'python', (1, 2))
print(tuple1)
print(type(tuple1))

# Chuyển đổi từ kiểu dữ liệu khác
list1 = [1, 2, [3, 4, 5]]
tuple2 = tuple(list1)
print(tuple2)
# a = tuple(i for i in range(10))
# b = tuple(range(10))
```

**Khả năng truy cập của Tuple**

```python
tuple3 = (1, 2, 3, 'python')

# Truy cập bằng chỉ số
print(tuple3[0])

# Truy cập bằng một khoảng chỉ số
print(tuple3[0:3])

# Truy cập bằng cách giải nén tuple
a, b, c, d = tuple3
e = tuple3
print(e)
print(a, b, c,d, sep = ", ")
```

**Vì Tuple là kiểu dữ liệu không thể thay đổi**

```python
tuple5 = (2, 3, 4, 5, 7, 8)
try:
    tuple5[0] = 6
except Exception as e:
    print(e)
```

```python
# Nối hai tuple
tuple6 = ('hit', 'python', 'công khai')
tuple7 = (2, 0, '26')
tuple_concate = tuple6 + tuple7
print(tuple_concate)
print(id(tuple6))

# Lặp lại tuple
tuple7 = tuple7 * 2
print(id(tuple7))
```

**1 SỐ PHƯƠNG THỨC**


```python
tuple8 = (1, 2, 3, 2, 4, 2, 5, 2.5)

count_2 = tuple8.count(2)
print(f"Số lần xuất hiện của 2: {count_2}")

pos_2 = tuple8.index(2)
print(f"Vị trí của phần tử '2': {pos_2}")
```
