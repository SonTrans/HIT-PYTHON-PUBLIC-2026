Bài 1: List – Quản lý điểm

- Nhập n điểm sinh viên vào một list.

- Yêu cầu:

    - Điểm phải từ 0 đến 10, nhập sai thì nhập lại.
    - Tính điểm trung bình.
    - Tìm điểm lớn nhất, nhỏ nhất mà không dùng max(), min().
    - Tạo danh sách các điểm lớn hơn điểm trung bình.
    - Kiểm tra có sinh viên đạt điểm 10 hay không.

Bài 2: Tuple – Quản lý sản phẩm

- Mỗi sản phẩm là một tuple:

    (ma_san_pham, ten_san_pham, don_gia, so_luong)

- Yêu cầu:

    - Nhập nhiều sản phẩm, mã sản phẩm không được trùng.
    - Đơn giá phải lớn hơn 0, số lượng không âm.
    - Tính thành tiền từng sản phẩm.
    - Tìm sản phẩm có giá trị lớn nhất.
    - In các sản phẩm sắp hết hàng (số lượng < 5).
    - Nhập mã sản phẩm và tìm kiếm thông tin.
    - Tính tổng giá trị kho.

Bài 3: Set – Quản lý câu lạc bộ

- Có ba set lưu sinh viên tham gia:

    python_club
    ai_club
    english_club

- Yêu cầu:

    - Tìm sinh viên tham gia ít nhất một câu lạc bộ.
    - Tìm sinh viên tham gia cả ba câu lạc bộ.
    - Tìm sinh viên chỉ tham gia Python.
    - Tìm sinh viên tham gia đúng một câu lạc bộ.
    - Tìm sinh viên tham gia ít nhất hai câu lạc bộ.
    - Nhập mã sinh viên và in các câu lạc bộ sinh viên đó tham gia.

Bài 4: String – Đăng ký tài khoản

- Viết chương trình đăng ký tài khoản.

- Yêu cầu:

    - Chuẩn hóa họ tên, không dùng title().
    - Tên đăng nhập phải:
        - Dài từ 6 đến 20 ký tự.
        - Bắt đầu bằng chữ cái.
        - Có ít nhất một chữ số.
        - Không chứa khoảng trắng hoặc ký tự đặc biệt.
    - Mật khẩu phải:
        - Có ít nhất 8 ký tự.
        - Có chữ hoa, chữ thường, chữ số và ký tự đặc biệt.
        - Không chứa khoảng trắng.
