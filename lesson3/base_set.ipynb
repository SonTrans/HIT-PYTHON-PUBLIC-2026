**Set là một tập hợp không có thứ tự gồm các phần tử duy nhất.**

```python
set1 = {1, 2, 3.1, 2, 'python', 'python'}
print(set1)
print(type(set1))
list1 = [1, 3, 4, 5, 4]
set2 = set(list1)
print(set2)
print(type(set2))
string1 = "CLB HIT"
set3 = set(string1)
print(set3)
```

```python
try:
    set4 = {[2, 3, 4], {5, 6, 7}}
    print(set4)
except Exception as e:
    print(e)
try:
    set5 = {'hehe', (1, 2, 3), 5, 7.3}
    print(set5)
except Exception as e:
    print(e)
```

```python
 # Không thể thay đổi trực tiếp giá trị của phần tử trong set
set6 = {6, 7 , 8}
try:
    set6[1] = 10     # Lỗi
    print(set6)
except Exception as e:
 print(e)
```

**Một số phương thức**

```python
set7 = {1, 4, 5, 'python', 'công khai'}

# Thêm một phần tử mới vào set bằng phương thức add(value)
set7.add(3)
print(set7)

print(len(set7))
# LƯU Ý: nếu thêm một phần tử đã tồn tại trong set thì phần tử đó sẽ không được thêm lại
set7.add(4)
print(set7)
```

```python
set7 = {1, 2, 3, 'python'}
set7.remove(1)
print(set7)
try:
    set7.remove(4)
    print(set7)
except Exception as e:
    print(e)
set7.discard(4)
print(set7)
set7.pop()
print(set7)
print(set7.clear())
```

```python
set8 = {'toi', 'yeu'}
set9 = {'HIT', 'toi'}
```

```python
result = set8.union(set9) #trả về các phần tử duy nhất của 2 set
print(result)
```

```python
result = set8.intersection(set9) #trả về các phần tử giống nhau của 2 set
print(result)
```

```python
result1 = set8.difference(set9)
result2 = set9.difference(set8)
print(result1)
print(result2)
```

```python
result = set8.symmetric_difference(set9)    # Kiểm tra các phần tử chỉ thuộc một trong hai set
print(result)
```

```python
result = set8.isdisjoint(set9)  # Kiểm tra hai set có phần tử chung hay không
print(result)
```

```python
result = set8.issubset(set9)    # Kiểm tra set thứ nhất có phải là tập con của set thứ hai hay không
print(result)
```
