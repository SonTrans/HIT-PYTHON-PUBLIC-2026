**1 SỐ PHƯƠNG THỨC**

```python
string1 = 'HIT PYTHON PUBLIC 2025'
string2 = string1.replace('2025', '2026')
print(f"chuỗi sau khi được thay thế phần tử: {string2}")
print(f"chuỗi ban đầu: {string1}")
```

```python
string3 = string2.split(" ")
print(string3)
```

```python
string4 = [' em', 'yêU', 'PytHon ']
kq = ' '.join(string4)
print(kq)
```

```python
print(f"chuỗi sau khi được viết in hoa upper(): {kq.upper()}")

print(f"chuỗi sau khi được viết in thường lower(): {kq.lower()}")

print(f"chuỗi sau khi được xóa khoảng trắng 2 đầu strip(): {kq.strip()}")
```

```python
string5 = "tôi@là kiều$lương1tâm "
print(f"title(): {string5.title()}")
```

```python
string6 = "Python123"

print(f"kiểm tra chuỗi chỉ chứa chữ cái không: {string6.isalpha()}")
print(f"kiểm tra chuỗi chỉ chứa chữ số không: {string6.isdigit()}")
print(f"kiểm tra chuỗi chỉ chứa chữ số và chữ cái không: {string6.isalnum()}")
print(f"kiểm tra chuỗi có bắt đầu bằng chuỗi con 'Python' không: {string6.startswith("Python")}")
print(f"kiểm tra chuỗi có kết thúc bằng chuỗi con '123' không: {string6.endswith("123")}")
```
